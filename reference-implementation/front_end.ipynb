```python
import numpy as np
import matplotlib.pyplot as plt
from implementation import NeuralNet
```

```python
data = np.genfromtxt("../data/train.csv", delimiter=",", skip_header=1, dtype=int)
answers = data[:,0]
images = data[:,1:]/255
```

```python
def plot_image(i):
    plt.imshow(images[i,:].reshape(28,28), cmap="gray")
    plt.title(answers[i])
    plt.show()
plot_image(0)
```

```python
net = NeuralNet([784, 30, 10])
net.evaluate((answers[:21_000], images[:21_000]))
```

```python
%%time
net.train(training_data=(answers[21_000:], images[21_000:]), iterations=(1, 1), batch_size=10, learning_rate=.1)
net.evaluate((answers[:21_000], images[:21_000]))
```
