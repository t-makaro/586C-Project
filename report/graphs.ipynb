```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
implementations = ['Python+NumPy', 'C++', 'Naive" CUDA']
x = np.arange(len(implementations))
times_60k = [2.92,13.29,110]
times_10k = [0.50,2.21,18]
colors = ['#1f77b4', 'orange', '#76B900']
width = 0.35

# Create a bar chart
plt.figure(figsize=(8, 6))
bars1 = plt.bar(x - width/2, times_60k, width, label='60k Images', color=colors)
bars2 = plt.bar(x + width/2, times_10k, width, label='10k Images', color=colors)

# Add labels and title
plt.xlabel('Implementation')
plt.ylabel('Execution Time (seconds)')
plt.xticks(x, implementations)
plt.title('Forward Propagation Execution Time Comparison')

# Save the plot to a file
plt.savefig('forward_prop_time_comparison.png')

# Add text labels on top of each bar
for bar, label in zip(bars1, ['60k'] * len(bars1)):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), label, ha='center', va='bottom')
for bar, label in zip(bars2, ['10k'] * len(bars2)):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), label, ha='center', va='bottom')

# Display the plot
plt.show()
```
