```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
implementations = ['Python+NumPy', 'C++', 'CUDA']
x = np.arange(len(implementations))
times_60k = [2.92,13.29,67.44]
times_10k = [0.50,2.21,11.49]
colors = ['#1f77b4', 'orange', '#76B900']
width = 0.35

# Create a bar chart
plt.figure(figsize=(8, 6))
bars1 = plt.bar(x - width/2, times_60k, width, label='60k Images', color=colors)
bars2 = plt.bar(x + width/2, times_10k, width, label='10k Images', color=colors)

# Add labels and title
plt.xlabel('Implementation')
plt.ylabel('Execution Time (seconds)')
plt.xticks(x, implementations)
plt.title('Forward Propagation Execution Time Comparison')

# Add text labels on top of each bar
for bar, label in zip(bars1, ['60k'] * len(bars1)):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), label, ha='center', va='bottom')
for bar, label in zip(bars2, ['10k'] * len(bars2)):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), label, ha='center', va='bottom')

# Save the plot to a file
plt.savefig('images/forward_prop_time_comparison.png')

# Display the plot
plt.show()
```

```python
implementations = ['Python+NumPy', 'C++', 'CUDA']
x = np.arange(len(implementations))
times_60k = [2.92,13.29,67.44]
times_10k = [0.50,2.21,11.49]
colors = ['#1f77b4', 'orange', '#76B900']
width = 0.35

# Create a bar chart
plt.figure(figsize=(8, 6))
bars1 = plt.bar(x - width/2, times_60k, width, label='60k Images', color=colors)
bars2 = plt.bar(x + width/2, times_10k, width, label='10k Images', color=colors)

# Add labels and title
plt.xlabel('Implementation')
plt.ylabel('Execution Time (seconds)')
plt.xticks(x, implementations)
plt.title('Backwards Propagation Execution Time Comparison')

# Add text labels on top of each bar
for bar, label in zip(bars1, ['60k'] * len(bars1)):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), label, ha='center', va='bottom')
for bar, label in zip(bars2, ['10k'] * len(bars2)):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), label, ha='center', va='bottom')

# Save the plot to a file
plt.savefig('images/backwards_prop_time_comparison.png')

# Display the plot
plt.show()
```

```python
# Data
batch_sizes = [1, 15, 30, 60, 120, 240, 480, 2400, 24000, 60000]
times_60k = [1, 0.9, 0.8, 0.7, 0.6, 0.5, 0.4, 0.3, 0.2, 0.1]

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(batch_sizes, times_60k, marker='o')

# Set labels, title, and x-axis scale
plt.xlabel('Batch Size')
plt.ylabel('Execution Time (ms)')
plt.title('Forward Propagation Batch Size Comparison')
plt.xscale('log')
plt.xticks(batch_sizes)

# Formatter for exact batch size values on x-axis
plt.gca().xaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x):,}'))

# Save the plot to a file
plt.savefig('images/forwards_batch_time_comparison.png')

# Display the plot
plt.show()
```
