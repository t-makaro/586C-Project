```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
implementations = ['Python+NumPy', 'C++', '"Stupid" CUDA', '"Naive" CUDA']
times = [7, 20, 20, 20]  # Example times in seconds for each implementation
colors = ['#1f77b4', 'orange', '#76B900', '#76B900']

# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(implementations, times, color=colors)

# Add labels and title
plt.xlabel('Implementation')
plt.ylabel('Execution Time (seconds)')
plt.title('Forward Propagation Execution Time Comparison (60k images)')

# Save the plot to a file
plt.savefig('forward_prop_time_comparison.png')

# Display the plot
plt.show()
```
